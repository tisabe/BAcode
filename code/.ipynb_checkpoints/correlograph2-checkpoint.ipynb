# Correlograph

This notebook demonstrates the computation of correlographs.

```python
import numpy as np
import numpy.random as random
import scipy.interpolate as sci
import matplotlib as matplotlib
import matplotlib.pyplot as p
import math
%matplotlib inline
import scipy.ndimage.filters as filters
matplotlib.rcParams['figure.figsize'] = (16.0, 5.0)
matplotlib.rc('font',**{'size' : 13})
```

Next, we generate some random data to apply the technique to. We will generate Npattern = 10 different patterns. These different patterns correspond to different probe positions, i.e. positions where the electron beam is pointed to on the specimen. In our simulation, these different probe positions are emulated by different rotation angles of the super cell and different translation vectors of the unit cell within that volume.

```python
NpixX = 128 # number of pixels along x
NpixY = 128 # number of pixels along y
Nk    = 100 # number of k-points in interpolated array
Nphi  = 50  # number of phi-points in interpolated array
Npattern = 3 # number of diffraction patterns to be simulated
dx = 1  # Pixel size along x
dy = 1  # pixel size along y
Nsignal = round(0.1*NpixX*NpixY)
sample = np.zeros((Npattern,NpixY,NpixX)) # scattering objects in real space
dp     = np.zeros((Npattern,NpixY,NpixX)) # intensities of diffraction patterns in (kx,ky) coordinates
dpPol  = np.zeros((Npattern,Nphi,Nk)) # interpolated diffraction intensities in polar coordinates
for ip in np.arange(Npattern):
    posX = (np.round((NpixX-1)*(random.random(Nsignal)))).astype('int')
    posY = (np.round((NpixY-1)*(random.random(Nsignal)))).astype('int')
    for jpix in range(0,Nsignal):
        sample[ip][posY[jpix]][posX[jpix]] = sample[ip][posY[jpix]][posX[jpix]]+1
    sample[ip] -= np.mean(sample[ip])+0.1/(NpixX*NpixY)

p.imshow(sample[0]); p.colorbar()
p.title("sample [0] - random dots")
```

Now we can simulate the diffraction patterns (dp) of our random arrangements of atoms. In the final code, these patterns should be computed using pyqstem. We then display the first 3 of these patterns:

```python
# the diffraction intensity is the amplitude squared of the FFT of object:
for ip in np.arange(Npattern):
    dp[ip] = np.abs(np.fft.fftshift(np.fft.fft2(sample[ip])))**2

# generate a reciprocal space grid:
kx1 = np.fft.fftshift(np.fft.fftfreq(NpixX,dx)) # 1D reciprocal space coordinate along x-direction 
ky1 = np.fft.fftshift(np.fft.fftfreq(NpixY,dy))
kX = np.tile(kx1.reshape(1,NpixX),(NpixY,1))
kY = np.tile(ky1.reshape(NpixY,1),(1,NpixX))

# display the 3 first diffraction patterns:   

if Npattern > 2:
    fig,(ax1,ax2,ax3)=p.subplots(1,3,figsize=(20,8))
    ax1.imshow(np.log(1+dp[0]),extent=[kx1[0],kx1[-1],ky1[0],ky1[-1]])
    p.title("dp [0]")  # does not work
    ax2.imshow(np.log(1+dp[1]),extent=[kx1[0],kx1[-1],ky1[0],ky1[-1]])
    p.title("dp [1]")  # does not work
    ax3.imshow(np.log(1+dp[1]),extent=[kx1[0],kx1[-1],ky1[0],ky1[-1]])
    p.title("dp [2]")
else:
    p.imshow(np.log(1+dp[0]),extent=[kx1[0],kx1[-1],ky1[0],ky1[-1]])
```

Next, we will compute the correlograph of these different simulated diffraction patterns. In a first step we must re-express thes diffraction patterns in polar coordinates on a regular grid of angles and radius, i.e. the x-axis becomes the radial distance from the undiffracted beam, and the y-axis becomes the angle. This involves 2D interpolation, of course.

```python
# generate kp and phi arrays 
import scipy.ndimage as scnd
kmax = np.max(np.sqrt(kX**2+kY**2))/(1.01*np.sqrt(2))
phi = np.arange(-np.pi,np.pi,2*np.pi/Nphi)  # does not include +pi
k   = np.linspace(0.0,kmax,Nk)              # includes kmax

# generate the grid on which we compute the interpolated data
k2,phi2 = np.meshgrid(k,phi)
kXi = k2*np.cos(phi2)
kYi = k2*np.sin(phi2)
for ip in np.arange(Npattern):
    interpSpline = sci.RectBivariateSpline(ky1, kx1, dp[ip]) 
    dpPol[ip] = interpSpline(kYi,kXi,grid=False) 
```

We have just converted the diffraction pattern from the (kx,ky)-space to (k,phi)-space. Lets display this now:

```python
p.imshow(np.log(1000+dpPol[0]),extent=[k[0],k[-1],phi[0],phi[-1]])
p.axis('auto')
p.axis('tight')
# p.imshow(np.log(1000+dpPol[0]))
# p.imshow(np.log(1+dp[0]))
p.colorbar()
kmax
```

The azimuthally averaged mean intensity can now very easily be computed for each pattern

```python
Ik=np.zeros((Npattern,Nk))
Iavg=np.zeros((1,Nk))
for ip in np.arange(Npattern):
    Ik[ip] = np.mean(dpPol[ip],axis=0)
Iavg = np.mean(Ik,axis=0)
p.plot(k,Ik.T)
p.title('azimuthally averaged diffraction intensities of all patterns')
Ik.shape
```

The normalized variance of these patterns is now computed as follows. 

```python
Ivar =np.zeros((1,Nk))
for ip in np.arange(Npattern):
    Ivar += (Ik[ip]-Iavg)**2
Ivar /= Npattern
Ivar /= Iavg**2
Ivar[0,0] = 0
p.plot(Ivar.T)
p.title('normalized variance of azimuthally averaged diffraction patterns')
```

Next, we can compute the correlograph of each of the diffraction patterns. For each diffraction pattern in (k,phi)-space we need to compute the cross-correlation along the phi-axis, i.e. the vertical (y-) axis.

```python
correlograph=np.zeros((Npattern,Nphi,Nk))
for ip in np.arange(Npattern):
    correlograph[ip] = np.real(np.fft.ifft(np.abs(np.fft.fft(dpPol[ip],axis=0))**2,axis=0))
    # Now we need to normalize this correlograph:
    correlograph[ip] /= np.tile(np.mean(correlograph[ip],axis=0), (Nphi, 1))
    
correlograph -= 1

if Npattern > 10:  # show the first two correlographs:
    fig,(ax1,ax2)=p.subplots(2,1,figsize=(16,10))
    ax1.imshow(correlograph[0],extent=[k[0],k[-1],2*np.pi,0])
    ax1.axis('auto'); ax1.axis('tight');
    ax2.imshow(correlograph[1],extent=[k[0],k[-1],2*np.pi,0])
    ax2.axis('auto'); ax2.axis('tight'); 
else: # show only one correlographs:
    p.imshow(correlograph[0],extent=[k[0],k[-1],2*np.pi,0])
    p.axis('auto'); p.axis('tight'); p.colorbar()   
```

Compute the mean correlograph as well as the variance of it.

```python
correlographAvg=np.mean(correlograph,axis=0)
p.imshow(correlographAvg,extent=[k[0],k[-1],2*np.pi,0])
p.axis('auto'); p.axis('tight'); p.colorbar()
p.title('average correlograph')
```

Now we can also compute the variance of the correlograph.

```python
correlographVar=np.zeros((Nphi,Nk))
for ip in np.arange(Npattern):
    correlographVar += (correlograph[ip]-correlographAvg)**2    
correlographVar /= correlographAvg**2
p.imshow(np.log(1+correlographVar),extent=[k[0],k[-1],2*np.pi,0])
p.axis('auto'); p.axis('tight'); p.colorbar()
p.title('Variance of correlographs')
```
