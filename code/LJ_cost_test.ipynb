```python
def cost_LJ_trunc(atoms, sigma, eps, rCut):
    arr = atoms.get_positions()
    shift = np.power(rCut, -6)
    r = pdist(arr)
    r = r[(r < rCut)]
    np.power(r/sigma, -6, out=r)
    return np.sum(r**2 - r - shift**2 + shift)*4*eps
def cost_LJ_trunc_mic(atoms, sigma, eps, rCut):
    r = atoms.get_all_distances(mic=False)
    r.flatten()
    r = r[(r != 0.0)]
    shift = np.power(rCut, -6)
    r = r[(r < rCut)]
    np.power(r/sigma, -6, out=r)
    return np.sum(r**2 - r - shift**2 + shift)*4*eps
```

```python
sigma = 2
eps = 1
rCut = 5.0
print(su.cost_LJ(struct1, sigma, eps))
print(cost_LJ_trunc(struct1, sigma, eps, rCut))
print(cost_LJ_trunc_mic(struct1, sigma, eps, rCut))
```
